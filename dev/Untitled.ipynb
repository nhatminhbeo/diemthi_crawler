```python
def clear_cache(timeout=60):
    """Clear the cookies and cache for the ChromeDriver instance."""
    # navigate to the settings page
    driver.get('chrome://settings/clearBrowserData')

    # wait for the button to appear
    wait = WebDriverWait(driver, timeout)
    wait.until(get_clear_browsing_button)

    # click the button to clear the cache
    get_clear_browsing_button(driver).click()

    # wait for the button to be gone before returning
    wait.until_not(get_clear_browsing_button)
```

```python
# get a search result with school id, last name, and page number
def getNext(id):
    # go to url
    driver.find_elements_by_id("txtkeyword")[0].clear()
    driver.find_elements_by_id("txtkeyword")[0].send_keys(id)

#     try:
#         element = driver.find_element_by_class_name('clearfix')
#         driver.execute_script("""var element = arguments[0]; element.parentNode.removeChild(element); """, element)
        
#         element = driver.find_element_by_class_name('text-danger')
#         driver.execute_script("""var element = arguments[0]; element.parentNode.removeChild(element); """, element)
#     finally:        
        
    WebDriverWait(driver, 10).until(EC.visibility_of_element_located((By.ID, "btnresult")))
    element = driver.find_element_by_id('btnresult')
    driver.execute_script("var element = arguments[0]; element.click()", element)
    #driver.find_elements_by_id("btnresult")[0].click()
    return BeautifulSoup(driver.page_source, 'lxml')
```

```python
def getStudent(soup):
        try:
            student = []
            soup = soup.find("tbody", {"id": "resultcontainer"})
            soup = soup.find("tr")
            soup = soup.findAll("td")
            for s in soup:
                data = s.text
                student.append(data)

            print(student)
            return student
        except:
            return None
```

```python
loop_time = time.time()
for i in range(159999, 160500):
    start_time = time.time()
    print(getStudent(getNext(str(i))))
    print("Done " + str(i) + " in " + str(time.time() - start_time) + "s")
loop_time = loop_time - time.time()
```
